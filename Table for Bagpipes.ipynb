This notebook aims at building a catalog of sources with fluxes
from a set of pictures taken with different filters 

```python
# Check python version
import sys
print(sys.version)
```

```python
from astropy.io import fits
```

```python
hdul = fits.open("UNCOVER_DR3_LW_SUPER_catalog.fits")
hdul.info()
```

```python
data = hdul[1].data
print(data.columns)
```

```python
id = data['id']
ra = data['ra']
dec = data['dec']
use_phot = data['use_phot']
f_f606w = data['f_f606w'] * 100 # conversion to muJy
f_f814w = data['f_f814w'] * 100 # conversion to muJy
f_f105w = data['f_f105w'] * 100 # conversion to muJy
f_f140w = data['f_f140w'] * 100 # conversion to muJy
f_f277w = data['f_f277w'] * 100 # conversion to muJy
f_f356w = data['f_f356w'] * 100 # conversion to muJy
f_f444w = data['f_f444w'] * 100 # conversion to muJy
e_f606w = data['e_f606w'] * 100 # conversion to muJy
e_f814w = data['e_f814w'] * 100 # conversion to muJy
e_f105w = data['e_f105w'] * 100 # conversion to muJy
e_f140w = data['e_f140w'] * 100 # conversion to muJy
e_f277w = data['e_f277w'] * 100 # conversion to muJy
e_f356w = data['e_f356w'] * 100 # conversion to muJy
e_f444w = data['e_f444w'] * 100 # conversion to muJy
```

```python
col_names = ["id", "ra", "dec", "use_phot",
             "f_f606w", "f_f814w", "f_f105w", "f_f140w","f_f277w", "f_f356w", "f_f444w", 
             "e_f606w", "e_f814w", "e_f105w", "e_f140w","e_f277w", "e_f356w", "e_f444w"]
print(len(data))
```

```python
import numpy as np
import pandas as pd

array = np.column_stack((id, ra, dec, use_phot,
                         f_f606w, f_f814w, f_f105w, f_f140w, f_f277w, f_f356w, f_f444w, 
                         e_f606w, e_f814w, e_f105w, e_f140w, e_f277w, e_f356w, e_f444w))
array = array[array[:, 3] == 1]
df = pd.DataFrame(array, columns=col_names)
print(len(df))
pd.set_option('display.precision', 3)
print("Table for bagpipes with flux and fluxerr")
print(" ")
print(df)
```

```python
# Save the catalog data to Excel

from astropy.table import Table

df.to_excel('Table_for_bagpipes.xlsx', index=False)
# Save the catalog data to .txt file
df.to_csv('Table_for_bagpipes.txt', sep='\t', na_rep="NaN", index=True, header=False)
```
