This notebook aims at building a catalog of sources with fluxes
from a set of pictures taken with different filters 

```python
# Check python version
import sys
print(sys.version)
```

```python
from astropy.io import fits
```

```python
hdul = fits.open("UNCOVER_DR3_LW_SUPER_catalog.fits")
hdul.info()
```

```python
data = hdul[1].data
print(data.columns)
```

```python
id = data['id']
x = data['x']
y = data['y']
ra = data['ra']
dec = data['dec']
kron_radius_circ = data['kron_radius_circ']
f_f606w = data['f_f606w']
f_f277w = data['f_f277w']
f_f356w = data['f_f356w']
f_f444w = data['f_f444w']
e_f606w = data['e_f606w']
e_f277w = data['e_f277w']
e_f356w = data['e_f356w']
e_f444w = data['e_f444w']
use_phot = data['use_phot']
```

```python
col_names = ["id", "x", "y", "ra", "dec", "kron_radius_circ",
             "f_f606w", "f_f277w", "f_f356w", "f_f444w", 
             "e_f606w", "e_f277w", "e_f356w", "e_f444w", "use_phot"]
```

```python
import numpy as np
import pandas as pd

array = np.column_stack((id, x, y, ra, dec, kron_radius_circ,
                         f_f606w, f_f277w, f_f356w, f_f444w, 
                         e_f606w, e_f277w, e_f356w, e_f444w, use_phot))
df = pd.DataFrame(array, columns=col_names) 
pd.set_option('display.precision', 2)
print("SUPER Catalog with flux and fluxerr")
print(" ")
print(df)
```

```python
# Save the catalog data to Excel
df.to_excel('SUPER_catalog.xlsx', index=False)
# Save the catalog data to .txt file
df.to_csv('SUPER_catalog.txt', sep='\t', na_rep="NaN", index=True, header=True)
```
